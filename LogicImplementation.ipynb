```python
from BtAssessmentLib.__BacktestModule import BTest
import math, pandas

b = BTest(ticker='SBIN',
          start_date='2023-07-28 09-15',
          end_date='2023-08-10 15-30',
          bar_interval='1min',
          quantity=1,
          capital=100000,
          stop_loss=7,
          target=5,
          change_bar_interval_at_start=True)
```

```python
df = b.get_df()
```

```python
df_dict = df.to_dict('index')
```

```python
position={}
curr_sl, curr_tp,signal = None,None, None
for k,v in df_dict.items():

    if signal is True:
        buyprice = v['OpenValue']
        if curr_sl is None:
            curr_sl = b.sl_cal(buyprice)
        if curr_tp is None:
            curr_tp = b.target_cal(buyprice)
        print(f"Short entry @ {v}")
        signal = False
        
    if all(not pandas.isna(x) for x in v.values()):
        curr_close = v['CloseValue']
        curr_ub = v['UB']
        # Entry Criteria
        if curr_close>curr_ub:        
            signal = True
            print(f'Short entry signal @ {v}')
        elif curr_tp is not None and curr_sl is not None:
            if curr_sl>curr_close:
                print(f"SL hit @ {v}")
            elif curr_tp<curr_close:
                print(f"target hit @ {v}")
            elif curr_close<curr_ub:
                print(f"Trend reversed @ {v}")
        # print(k,v)
```
